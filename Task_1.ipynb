# Extraction of top 10 countries with respect to total population

```python
#Filter data for total population
total_population_data=df[df['Series Code']=='SP.POP.TOTL']

#Sort data based on total population of 2022
total_population_sorted= total_population_data.sort_values(by="2022",ascending=False)

#Get the top 10 countries with highest total population of 2022
total_top_10_countries=total_population_sorted.head(10)

print("Top 10 countries with total population\n")
print(total_top_10_countries[['Country Code']])
```

# Bar Plot

# Top 10 countries of total population in year 2022 and 2016

```python
#create a bar plot
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2022",y="Country Code",data=total_top_10_countries,palette="cool")
plt.title("Top 10 countries for total population of 2022")
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2016",y="Country Code",data=total_top_10_countries,palette="cool")
plt.title("Top 10 countries for total population of 2022")
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

# Extraction of top 10 countries with highest male population

```python
#Filter data for total population
male_population_data=df[df['Series Code']=='SP.POP.TOTL.MA.IN']

male_population_sorted= male_population_data.sort_values(by="2022",ascending=False)

male_top_10_countries=male_population_sorted.head(10)

print("Top 10 countries with highest male population\n")
print(male_top_10_countries[['Country Code']])
```

# Extraction of top 10 countries with highest female population

```python
#Filter data for total population
female_population_data=df[df['Series Code']=='SP.POP.TOTL.FE.IN']

female_population_sorted= female_population_data.sort_values(by="2022",ascending=False)

female_top_10_countries=male_population_sorted.head(10)

print("Top 10 countries with highest female population\n")
print(female_top_10_countries[['Country Code']])
```

# Top 10 countries with highest male and female population

```python
#create a bar plot
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2022",y="Country Code",data=male_top_10_countries,palette="cool")
plt.title("Top 10 countries for total population of 2022")
plt.xlabel("Male population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2016",y="Country Code",data=female_top_10_countries,palette="cool")
plt.title("Top 10 countries for total population of 2022")
plt.xlabel("Female population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

# Stacked Bar Plot

# Top 10 countries with male and female population

```python
#merge male and female population data on 'Country Code'
merge_data=pd.merge(male_population_data,female_population_data,on="Country Code",suffixes=("_male","_female"))
```

```python
#merged data
#calculate the total population for each country (male + female)
merge_data["Total population"] = merge_data["2022_male"] + merge_data ["2022_female"]
```

```python
merge_data.head()
```

```python
#sort data based on total population in descending order
sorted_data = merge_data.sort_values(by="Total population", ascending=False)
```

## Select the top 10 countries with highest population

```python
top_10_countries = sorted_data.head(10)
```

```python
#create the staked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code",y="2022_female",data=top_10_countries,color="blue",label="female Population")
sns.barplot(x="Country Code",y="2022_male",data=top_10_countries,bottom=top_10_countries["2022_female"],color="yellow",label="Male Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```

## Botton 10 countries with male and female population 2022

```python
#select the top 10 countries with the highest total population
bottom_10_countries = sorted_data.tail(10)
```

```python
# Create the stacked bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x="Country Code", y="2022_female", data=bottom_10_countries, color="blue", label="Female Population")
sns.barplot(x="Country Code", y="2022_male", data=bottom_10_countries, bottom=bottom_10_countries["2022_female"], color="yellow", label="Male Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```
